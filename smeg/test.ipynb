```python
def return_ele(soup, texte):
    try:
        return soup.find('div', text=re.compile(texte)).next_sibling.next_sibling.text.strip()
    except:
        return ''
```

```python
urls1 = pd.read_excel('smeg_url_update.xlsx')
list_urls = []

for idx, row in urls1.iterrows():
    list_urls.append({
        'url': row['url'],
        'cat1': row['cat1'],
        'cat2': row['cat2'],
        'cat3': row['cat3'],
    })

df =  pd.read_excel('smeg_product_model.xlsx')
for i, url2 in enumerate(list_urls):
    url = url2['url']
    cat1 = url2['cat1']
    cat2 = url2['cat2']
    cat3 = url2['cat3']
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    cookies = {'session': '134-8225175-0355220'}
    r = requests.get(url, headers=headers, cookies=cookies)
    soup2 = BeautifulSoup(r.content, "html.parser")
    
    print('Count: ', i)
    print('URL: ', url)
    
    name = soup2.find('h1', {'class': 'main__title'}).text.strip()
    sku = soup2.find('span', {'class': 'code'}).text.strip()
    free_colors = return_ele(soup2, 'Colour') 
    motor_power = return_ele(soup2, 'Motor power')
    variable_speed = return_ele(soup2, 'Variable speed')
    blades_material = return_ele(soup2, 'Blades material')
    finishing = return_ele(soup2, 'Finishing')
    
    category = return_ele(soup2, 'Category')
    aesthetic = return_ele(soup2, 'Aesthetic')
    cleaning_system = return_ele(soup2, 'Cleaning system')
    
    
    images = soup2.find_all('img', {'class': 'img-fluid'} )
    len(images)
    list_images = [ img['src'] for img in images]
    base_image = list_images[0]
    add_images = ','.join(list_images[1:])
    description = soup2.find('div', {'class': 'product-details__infos'}).text.replace('\n', '').replace('                                                                    ', '-').strip()
    data = {
        'sku': sku,
        'name': name,
        'url_link': url,
        'free_colors': free_colors,
        'motor_power': motor_power,
        'variable_speed': variable_speed,
        'blades_material': blades_material,
        'category': category,
        'aesthetic': '',
        'cleaning_system': cleaning_system,
        'finishing': finishing,
        'base_image': base_image,
        'add_images': add_images,
        'description': description,
        'cat1': cat1,
        'cat2': cat2,
        'cat3': cat3,
    }
    df1 = pd.DataFrame([data])
    df = pd.concat([df, df1], ignore_index=True)
    df.to_excel('smeg_update_product.xlsx')
```
