```python
def return_size(text):
    
    regex_h = re.compile(r'الإرتفاع=(?P<hight>\s*\d+.?\d+\s*)')
    regex_w = re.compile(r'العرض=(?P<width>\s*\d+.?\d+\s*)')
    regex_l1 = re.compile(r'العمق=(?P<lenght>\s*\d+.?\d+\s*)')
    regex_l2 = re.compile(r'العمق = (?P<lenght>\s*\d+.?\d+\s*)')
    p1 = re.search(regex_h, text)
    p2 = re.search(regex_w, text)
    p3 = re.search(regex_l1, text)
    p4 = re.search(regex_l2, text)
    if p1 != None:
        hight = p1['hight']
    if p2 != None:
        width = p2['width']
    if p3 != None:
        lenght = p3['lenght']
    if p4 != None:
        lenght = p4['lenght'] 
    return width, lenght, hight
```
