```python
def clear_size(text):
    regex = re.compile(r'Dimensions In Inches:w- (?P<width>\s*\d*.?\d+\s*) x d- (?P<lenght>\s*\d*.?\d+\s*) x h- (?P<hight>\s*\d*.?\d+\s*)')
    p = re.match(regex, text)
    
    if p != None:   
        width = p['width']   
        lenght = p['lenght'] 
        height = p['hight'] 
        print('W:', width)
        print('H:', height)
        print('L:', lenght)
        return float(width) , float(lenght) , float(height) 
    else:
        width = 0  
        lenght = 0
        height = 0
        return width, lenght, height
```
