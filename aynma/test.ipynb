```python
list_urls = pd.read_csv('aynma_url_update1.csv')
list_toto = []

for indx, row in list_urls.iterrows():
    list_toto.append(
        {'url': row['url'], 'cat1': row['cat1'], 'cat2': row['cat2'], 'cat3': row['cat3']}
    )
    
df = pd.read_excel('aynma_product_model.xlsx')

for i, url1 in enumerate(list_toto):
    print('Count: ', i)
    url = url1['url']
    cat1 = url1['cat1']
    cat2 = url1['cat2']
    cat3 = url1['cat3']
    
    print('URL: ', url)
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    cookies = {'session': '134-8225175-0355220'}
    r = requests.get(url, headers=headers, cookies=cookies)
    soup = BeautifulSoup(r.content, 'html.parser')
    
    try:
        name = soup.find('h1', {'class': 'product-details__title brand-title'}).text.strip()
    except:
        continue

    special_price = soup.find('span', id=re.compile('^price-after_') ).text.strip()
    price = soup.find('span', id=re.compile('^price-before_') ).text.strip()
    if price == '':
        price = soup.find('span', {'class': 'product-price'}).text.strip()
        special_price = ''
    try:
        sku = soup.find('b', text=re.compile('SKU:')).next_sibling.text.strip()
    except:
        sku = ''
    description =  soup.find('div', {'class': 'product-detials__desc'}).text.strip()
    try:
        is_in_stock = soup.find('span', {'class': 'not-available-large notify'}).text.strip()
    except:
        is_in_stock = '1'
    images = soup.find('div', {'id': 'sp-slider-cont'}).find_all('img')
    len(images)
    list_images = [img['src'] for img in images]
    base_image = list_images[0]
    add_images = ','.join(list_images[1:])
    
    
    data = {
        'sku': sku,
        'name': name,
        'link_url': url,
        'price': price,
        'special_price': special_price,
        'is_in_stock': is_in_stock,
        'description': description,
        'base_image': base_image,
        'add_images': add_images,
        'cat1': cat1,
        'cat2': cat2,
        'cat3': cat3,
        
    }
    
    df1 = pd.DataFrame([data])
    df = pd.concat([df, df1], ignore_index=True)
    df.to_csv('Aynbma_product1.csv')

```
