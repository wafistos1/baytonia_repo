```python
def filter_data_size_regex18(text):
    m = re.match(r'(\s*\d+.?\d+\s*) . . (\s*\d+.?\d+\s*) . . (\s*\d+.?\d+\s*) . .', text)
    width = m.group(1)
    lenght = m.group(2)
    hight = m.group(3)
    return width, lenght, hight
```

```python
def filter_data_size_regex1(text):
    
    width = 0.0
    lenght = 0.0
    hight = 0.0
#     text1 = text.split('x')
#     print('split text', text1)
    p = cas1_regex.match(text)
    print(p)
    try:
        width = float(p.group(2)) * 2.54
        print('width', width)
    except:
        width = p.group(2) * 2.54
        print('width', width)

#         print(x)
    try:
        lenght = float(p.group(3)) * 2.54
        print('lenght', lenght)
    except:
        lenght = p.group(3) * 2.54
        print('lenght', lenght)


#         print(x)
    try:
        
        hight = float(p.group(4))* 2.54 
        print('hight', hight)
    except:
        hight = p.group(4) * 2.54
        print('hight', hight)
    print('W:', width)
    print('L:', lenght)
    print('H:', hight)
    return width, lenght, hight
```

```python
def return_centimers(inches):
    
    pattern = re.compile(r'(\w+) \((\d+\.\d+ )\)')
    result = None
    if inches == None or inches == 0 or inches == '':
        print('elif cent')

        return 0
    elif pattern.search(inches) != None:
        print('elif cent')
        toto = pattern.search(inches)

        return toto.group(2)

    else:
        try:
            inches = inches.replace('"', '')
            print('j ai remplacer', inches)
            try:
                inches = int(inches)
                result = inches * 2.54
                result = f"{result:.2f}"
                return result
            except:
               
                inches = float(inches)
                print('foat', inches)
                result = inches * 2.54
                result = f"{result:.2f}"
                print('result',result)
                return result
        except:
            pass
```
