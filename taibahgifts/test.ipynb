```python

def join_images(table):
    return 'https://taibahgifts.com/pub/media/catalog/product/cache/dc3baba06599bac140722d2fe215553c/' +'/'.join(table)

df = pd.read_excel('taibahgifts_models.xlsx')

toto = pd.read_excel('Taibahgift_url.xlsx')
list_urls = []
for index, row in toto.iterrows():
    list_urls.append({'url': row['url'], 'cat1': row['cat1'], 'cat2': row['cat2']})

name_excel = 'Taibagifts_test.xlsx'

for i, url in enumerate(list_urls[169: ]):
    print('Count: ', i)
    
    cat1 = url['cat1']
    cat2 = url['cat2']
    url1 = url['url']
    print('URL: ', url1)
    driver.get(url1)
    time.sleep(5)
    def return_details(name, soup):
        try:
            return soup.find('strong', text=re.compile(name)).next_sibling
        except:
            return ''
    tex = driver.find_element_by_tag_name('body').get_attribute('innerHTML')
    soup = BeautifulSoup(tex, 'html.parser')
    
    name = soup.find('span', {'class': 'base'}).text.strip()

    is_in_stock = soup.find('span', text='التوفر:').next_element.next_element.next_element.text.strip()
    try:
        short_description = soup.find('div', {'class': 'product attribute overview'}).text.strip()
    except:
        short_description = ''
    try:
        description = soup.find('div', {'id': 'description'}).text.strip()
    except:
        try:
            
            description = soup.find('div', {'id': 'additional'}).text.strip()
        except:
         description = ''
    size = return_details('الحجم', soup)
    type_ = return_details('الفئة', soup)
    products_size = return_details('المقاس', soup)
    try:
        images = driver.find_elements_by_xpath('//img[@class="fotorama__img"]')
        len(images)
        list_images = []
        for img in images:
            list_images.append(join_images(img.get_attribute('src').split('/')[9:]))
        
        base_image = list_images[0]
        additionnel_images = ','.join(list_images[1:])
    except:
        base_image = soup.find('img', {'class': 'fotorama__img'})['src']
        additionnel_images = ''
    tt = driver.find_elements_by_class_name('swatch-option')
    if len(tt) != 0:
    
        for t in tt:
            try:
                t.click()
                time.sleep(1)
            except:
                continue   
            try:
                tex = driver.find_element_by_tag_name('body').get_attribute('innerHTML')
                soup = BeautifulSoup(tex, 'html.parser')
                price = soup.find('span', {'data-price-type': 'oldPrice'}).text.replace('ر.س', '').strip()
                special_price = soup.find('span', {'data-price-type': 'finalPrice'}).text.replace('ر.س', '').strip()
                products_size = t.get_attribute('aria-label')
                print('Products_size: ', products_size)
            except:
                price = soup.find('span', {'data-price-type': 'finalPrice'}).text.replace('ر.س', '').strip()
                special_price = '0'
            data = {
                'name': name,
                'cat1': cat1,
                'cat2': cat2,
                'link_url': url1,
                'price': price,
                'special_price': special_price,
                'is_in_stock': is_in_stock,
                'short_description': short_description,
                'description': description,
                'size': size,
                'type_': type_,
                'products_size': products_size,
                'base_image': base_image,
                'additionnel_images': additionnel_images,
            }
            
            df1 = pd.DataFrame([data])
            df = pd.concat([df, df1], ignore_index=True)
            df.to_excel(name_excel)
    else:
        tex = driver.find_element_by_tag_name('body').get_attribute('innerHTML')
        soup = BeautifulSoup(tex, 'html.parser')
        try:
            price = soup.find('span', {'data-price-type': 'oldPrice'}).text.replace('ر.س', '').strip()
            special_price = soup.find('span', {'data-price-type': 'finalPrice'}).text.replace('ر.س', '').strip()
        except:
            try:
                price = soup.find('span', {'data-price-type': 'finalPrice'}).text.replace('ر.س', '').strip()
                special_price = '0'
            except:
                continue
        data = {
            'name': name,
            'cat1': cat1,
            'cat2': cat2,
            'link_url': url1,
            'price': price,
            'special_price': special_price,
            'is_in_stock': is_in_stock,
            'short_description': short_description,
            'description': description,
            'size': size,
            'type_': type_,
            'products_size': products_size,
            'base_image': base_image,
            'additionnel_images': additionnel_images,
        }
        
        df1 = pd.DataFrame([data])
        df = pd.concat([df, df1], ignore_index=True)
        df.to_excel(name_excel)
```
