```python
def extract_ele(name, soup):
    try:
        return soup.find('strong', text=re.compile(name)).next_element.next_element.strip()
    except:
        return ''
```

```python
from dataclasses import dataclass, field

@dataclass
class afkarItem:
    """Class for keeping track of an item in inventory."""
    name: str = field(init=False)
    details: str= field(init=False)
    price: str= field(init=False)
    special_price: str= field(init=False)
    description: str= field(init=False)
    raw_materials: str= field(init=False)
    qualite_materials: str= field(init=False)
    color_exter_cadre: str= field(init=False)
    product_size: str= field(init=False)
    cat1: str= field(init=False)
    cat2: str= field(init=False)
    cat3: str= field(init=False)
    base_image: str= field(init=False)
    add_images: str= field(init=False)
    l_urls: list = field(default_factory=list)
    data: list = field(default_factory=list)
    
    def __post_init__(self):
        self.add_images = ''
    
    def get_find_element(self):
        self.name = soup.find('h1', {'class': 'product-details__title'}).text.strip()
        self.details = soup.find('h4', {'class': 'product-details__subtitle'}).text.strip()
        self.price = soup.find('span', {'class': 'price-before'}).text.strip()
        self.special_price = soup.find('span', {'class': 'price-after'}).text.strip()
        self.description = soup.find('div', {'class': 'product-detials__desc'}).text.strip()
        self.raw_materials = extract_ele('مادة الصنع', soup)
        self.qualite_materials = extract_ele('جودة المواد', soup)
        self.color_exter_cadre = extract_ele('لون الاطار الخارجي', soup)
        self.product_size = soup.find('strong', text=re.compile('المقاسات')).next_element.next_element.text.strip()
        self.base_image = None
        self.add_images = None
        self.cat1 = None
        self.cat2 = None
        self.cat3 = None
        
    
    def get_data(self, url):
        count = 1
        url = url + '?page=1'
        while True:
            print('Url:', url)
            headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
            cookies = {'session': '134-8225175-0355220'}
            r = requests.get(url, headers=headers, cookies=cookies)
            soup = BeautifulSoup(r.content, "html.parser")
            time.sleep(1)
            products = soup.find_all('div', {'class': 'product'})
            print(len(products))
            if len(products) == 0:
                break
            time.sleep(1)
            for toto in products:
                self.l_urls.append(toto.find('a')['href'])
            count += 1
            url = url.split('?')[0] + f'?page={count}'
        print('Len products', len(self.l_urls))
        return self.l_urls
    
    
    def scrape_url(self, list_urls: list):
        cat1 = list_urls['cat1']
        cat2 = list_urls['cat2']
        cat3 = list_urls['cat3']
        url = list_urls['url']
        data = []
        print(cat1, cat2, cat3)
        while True:
            urls_list = self.get_data(url)
            
            for toto in urls_list:
                self.data.append({
                'url':toto,
                'cat1': cat1,
                'cat2': cat2,
                'cat3': cat3,
                })
            print( f'Scrape done .')
            return self.data
    
    
    def scrape_detail(self, list_urls):
        pass
        
```

```python
def select_text(text):
    soup = BeautifulSoup(text, 'html.parser')
    return soup.find('input')['data-field_value']
```

```python
def scrap_data(driver):
    r = driver.find_element_by_tag_name('body').get_attribute('innerHTML')
    soup = BeautifulSoup(r, 'html.parser')
    # Details
    name = soup.find('h1', {'class': 'product-details__title'}).text.strip()
    details = soup.find('h4', {'class': 'product-details__subtitle'}).text.strip()
    price = soup.find('span', {'class': 'price-before'}).text.strip()
    special_price = soup.find('span', {'class': 'price-after'}).text.strip()
    description = soup.find('div', {'class': 'product-detials__desc'}).text.strip()
    raw_materials = extract_ele('مادة الصنع', soup)
    qualite_materials = extract_ele('جودة المواد', soup)
    color_exter_cadre = extract_ele('لون الاطار الخارجي', soup)
    product_size = soup.find('strong', text=re.compile('المقاسات')).next_element.next_element.text.strip()
    special_price = ''
    data = {
        'name': name,
        'details': details,
        'price': price,
        'special_price': special_price,
        'description': description,
        'raw_materials': raw_materials,
        'qualite_materials': qualite_materials,
        'color_exter_cadre': color_exter_cadre,
        'product_size': product_size,
        'special_price': special_price,
    }
    return data
```
