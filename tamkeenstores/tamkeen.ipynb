```python
def scrape_all_categories(pat):
    list_urls_categories = []
    
    print('URL:', pat['url'])
    # proxy = '64.124.38.140:8080'
#     for tt in PROXY_LIST:
#         print('proxy', tt)
#         try:
#             r = requests.get(pat['url'], proxies={'http': tt, 'https': tt}, timeout=10 )
#             if r.status_code == 200:
#                 print(r.status_code)
#                 break
#         except:
#             pass
        
        
    r = requests.get(pat['url'])
    soup  = BeautifulSoup(r.text, 'html.parser')
    
    urls = soup.find_all('div', {'class': 'product-item__body pb-xl-2'})
    len(urls)
    cat1 = pat['cat1']
    cat2 = pat['cat2']
    cat3 = pat['cat3']
    for url in urls:
        toto =  url.find('a')['href']
        if not 'https://' in toto:
            toto = 'https://tamkeenstores.com.sa/product/' + toto

        list_urls_categories.append({ 'url': toto,
                                        'cat1': cat1,
                                        'cat2': cat2,
                                        'cat3': cat3,
                                    })
    print('Len:', len(list_urls_categories))
    return list_urls_categories
```
