```python
from pandas import ExcelWriter

# writer = ExcelWriter('Masterfile.xlsx')

list_pro = []
list_images = []
list_input_title = []
list_input_price = []
list_input_images1 = []
list_input_images2 = []
list_input_images3 = []
list_input_price_was = []
list_input_model = []
list_input_size = []
list_input_color = []
list_input_mater = []
list_input_made_in = []
list_input_video = []
list_input_category = []
def try_product(product):
    pro = ''
    try:
        pro = product
    except:
        pass
    return pro


count = 2

def clear_data(data, tableau):
    value_return = ''
    for t in tableau:
        if data in t:
            value_return = t.replace(str(data), '')
    return value_return        


table = []
for url in list_products:
    #print(url)
    driver.get(url)
    product  = Product()
    
    product.title = driver.find_element_by_css_selector('h1').text
    try:
        product.price = driver.find_element_by_class_name('priceNow').text[5:10]
    except:
        product.price = ''
    model = driver.find_element_by_class_name('modelNumber').text[14:]
    category  = driver.find_element_by_class_name('loqmag-0').text
    try:
        product.price_was = driver.find_element_by_class_name('priceWas').text[5:10]
    except:
        product.price_was = 0
        
    
    #extract المواصفات
    details = driver.find_element_by_class_name('sc-1oxmzfx-0')
    parse = details.text
    parse2 = parse.replace('\n', '-')
    parse2 = parse2.split('-')
    print(len(parse2))
    #print(parse2)
    clean_list = parse2[1:]
    
    color = clear_data('اسم اللون', clean_list)
    mater = clear_data('المادة', clean_list)
    video = clear_data('الفيديو', clean_list)
    size = clear_data('الحجم', clean_list)
    model = clear_data('رقم الموديل', clean_list)
    made_in = clear_data('بلد المنشأ', clean_list)
    #print(color, mater, video, size, model)
    print(clean_list)

    #extract images
    images = driver.find_elements_by_css_selector('img')
    my_regex_pattern = r"(^https://z.nooncdn.com/products/tr:n-t_400/)|(^\/{1,2})"
    regex = re.compile(my_regex_pattern)
    for tit in images:
        
        try:
            
            my_match = regex.match(tit.get_attribute('src'))
            if my_match != None:
                list_images.append(tit.get_attribute('src'))
                print(tit.get_attribute('src'))
        except:
            pass
    print('list image ', list_images)

    list_input_title.append(product.title)
    list_input_price.append(product.price)
    list_input_price_was.append(product.price_was)
    list_input_model.append(model)
    list_input_mater.append(mater)
    list_input_color.append(color)
    list_input_size.append(size)
    list_input_video.append(video)
    list_input_made_in.append(made_in)
    list_input_category.append(category)

    try:
        list_input_images1.append(list_images[0])
    except:
        list_input_images1.append('')
    
    try:
        list_input_images2.append(list_images[1])
    except:
        list_input_images2.append('')
    
    try:
        list_input_images3.append(list_images[2])
    except:
        list_input_images3.append('')

    #df['Images'] = list_images
    
    df = pd.DataFrame({'Title': list_input_title,
                       'Price': list_input_price, 
                       "Price was":list_input_price_was,  
                       'Model': list_input_model, 
                       'Color': list_input_color, 
                       'Mater': list_input_mater,
                       'Made in ': list_input_made_in,
                       'Video': list_input_video,
                       'size': list_input_size,
                       'Category': list_input_category,
                       'Images1': list_input_images1,
                       'Images2': list_input_images2,
                       'Images3': list_input_images3,
                       
                      })
    df.to_excel( 'resultNoon.xlsx', index = True)
    list_images = []
    count += 1
    time.sleep(2)
    print('count ', count)
```
